```python
# LÍNEA EVALUABLE, NO RENOMBRAR LAS VARIABLES
import pyspark.sql.functions as F
import numpy as np
from pyspark.sql.types import *
from pyspark.sql import Window
from functools import reduce


USvideosDF = spark.read\
                  .option("inferSchema", "false")\
                  .option("quote", "\"")\
                  .option("escape", "\"")\
                  .option("header", "true")\
                  .csv('gs://clase1_ntic/data/youtube_USvideos.csv', mode = "DROPMALFORMED")\
                  .withColumn('pais', F.lit('EEUU'))\
                  .drop('description') 

CAvideosDF = spark.read\
                  .option("inferSchema", "false")\
                  .option("quote", "\"")\
                  .option("escape", "\"")\
                  .option("header", "true")\
                  .csv('gs://clase1_ntic/data/youtube_CAvideos.csv', mode = "DROPMALFORMED")\
                  .withColumn('pais', F.lit('CA'))\
                  .drop('description') 

GBvideosDF = spark.read\
                  .option("inferSchema", "false")\
                  .option("quote", "\"")\
                  .option("escape", "\"")\
                  .option("header", "true")\
                  .csv('gs://clase1_ntic/data/youtube_GBvideos.csv', mode = "DROPMALFORMED")\
                  .withColumn('pais', F.lit('GB'))\
                  .drop('description') 

videosRawDF = USvideosDF.union(CAvideosDF).union(GBvideosDF) 
# YOUR CODE HERE
# raise NotImplementedError()
```

* Las columnas `trending_date` y `publish_time` son en realidad de tipo fecha y de tipo timestamp (instante de tiempo), respectivamente, que Spark debería procesar como tales. Partiendo de `vidosRawDF`, **reemplaza** cada una de estas columnas por su versión convertida al tipo de dato correcto en cada caso, utilizando `withColumn` con el mismo nombre de la columna existente. Para las columnas de tipo fecha y timestamp, el nuevo valor de la columna viene dado por el siguiente código:

```python
# No olvides los imports que necesites...
# LÍNEAS EVALUABLES, NO RENOMBRAR LAS VARIABLES
videosDF = videosRawDF.withColumn('trending_date', F.to_date("trending_date", "yy.dd.MM"))\
                      .withColumn('publish_time', F.from_unixtime(F.unix_timestamp('publish_time',\
                                                   "yyyy-MM-dd'T'HH:mm:ss")).cast("timestamp"))\
                      .withColumn('likes', F.col('likes').cast(IntegerType()))\
                      .withColumn('dislikes', F.col('dislikes').cast(IntegerType()))\
                      .withColumn('comment_count', F.col('comment_count').cast(IntegerType()))\
                      .withColumn('comments_disabled', F.col('comments_disabled')\
                                                       .cast(BooleanType()))\
                      .dropna()\
                      .cache() 
# YOUR CODE HERE
# raise NotImplementedError()
```

* Calcular un nuevo DF llamado `diasViralCategoriaPaisDF` con tantas filas como **categorías** y con tantas columnas como **países** (es decir, 3 más la columna de categorías), que contenga en cada celda el número **medio** de días que un vídeo de cada categoría permanece siendo viral en ese país. * Por último, calcular otro DF llamado `videosPorDiasemanaDF` con tantas filas como **categorías** y tantas columnas como **días de la semana**, de forma que cada celda contenga el **número de vídeos** que se han publicado en esa categoría ese día de la semana. Tras el cálculo, **renombrar las columnas** para que sus nombres sean "Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo" en lugar de los números enteros del 1 al 7.

```python
# LÍNEA EVALUABLE, NO RENOMBRAR VARIABLES
# import.....
ventanaVideoId = Window().partitionBy("video_id") 
videosDiasViralDF = videosExtraInfoDF\
                    .withColumn("trending_date_max",F.max("trending_date").over(ventanaVideoId))\
                    .withColumn("trending_date_min",F.min("trending_date").over(ventanaVideoId))\
                    .withColumn("dias_viral",F.datediff('trending_date_max','trending_date_min'))\
                    .drop('trending_date_max','trending_date_min')\
                    .dropDuplicates(["video_id"])\
                    .cache() 

diasViralCategoriaPaisDF = videosDiasViralDF.groupBy("category_id")\
                            .pivot("pais").agg(F.mean("dias_viral"))  

videosPorDiasemanaDF = videosDiasViralDF.groupBy("category_id").pivot("diasemana").count()
oldColumns = videosPorDiasemanaDF.columns
newColumns = ["category_id", "Lunes", "Martes", "Miércoles", "Jueves", "Viernes", 
              "Sábado", "Domingo"]

videosPorDiasemanaDF = reduce(lambda data, i: data\
                              .withColumnRenamed(oldColumns[i], newColumns[i]), 
                              range(len(oldColumns)), videosPorDiasemanaDF) 


videosPorDiasemana = videosPorDiasemanaDF.sort("category_id").collect() 
# YOUR CODE HERE
# raise NotImplementedError()
```

Partiendo de `videosDiasViralDF`, donde cada vídeo solo aparece una vez, añadir las siguientes columnas:

* Una columna entera `ocurrencias_music` con el **número de ocurrencias** de la cadena **"music"** como **subcadena de cualquiera de los tags**. No es necesario que el tag sea exactamente igual a "music" sino que lo contenga como subcadena (recordemos que el DF de partida ya tiene todos los tags pasados a minúscula). Para ello, implementa una **UDF** `subtag_music_UDF` que devuelva `IntegerType()` y que envuelva a una función convencional de python llamada `subcadena_en_vector(tags)`. Esta última debe recibir como argumento una lista de strings, y comprobar cuántos elementos del vector contienen como subcadena a la palabra "music". Prueba su funcionamiento con la lista de tags `["a life in music", "music for life", "bso", "hans zimmer"]` que debería devolver 2. Finalmente invoca a `subtag_music_UDF` dentro de `withColumn` sobre la columna `tags`.

* Una columna `ocurrencia_media` de números reales con el número **medio** de apariciones de la palabra "music" como subcadena de tags en los vídeos similares al de la fila actual, entendiendo similares como aquellos del **mismo país y la misma categoría**. No debe utilizarse JOIN sino funciones de ventana.

* Una nueva columna `diff_porcentaje` que indique, en tanto por ciento, en qué medida el número de aparticiones de "music" está por encima o por debajo de la media de los vídeos de su misma categoría y país. Debe calcularse mediante **operaciones aritméticas entre columnas**, sin usar la función `F.when`, como la diferencia entre el número de aparticiones en el vídeo actual menos las apariciones medias de su país y categoría, dividido entre este último y multiplicado por 100.

El resultado de todas estas transformaciones debe quedar almacenado en la variable `videosOcurrenciasMusicDF`.

```python
# LÍNEA EVALUABLE, NO RENOMBRAR VARIABLES
# imports necesarios..........
def subcadena_en_vector(tags):
    return (sum([1 for c in tags if "music" in c])) 
subtag_music_UDF = F.udf(subcadena_en_vector, IntegerType())
subtag_music = videosDiasViralDF.withColumn("ocurrencias_music", subtag_music_UDF(F.col("tags"))) 

ventanaCategoriaPais = Window().partitionBy("pais","category_id") 
videosOcurrenciasMusicDF = subtag_music.withColumn('ocurrencia_media',\
                                            F.mean('ocurrencias_music')\
                                            .over(ventanaCategoriaPais))\
                                        .withColumn('diff_porcentaje',\
   ((F.col('ocurrencias_music')-F.col('ocurrencia_media'))/F.col('ocurrencia_media'))*100) 
# YOUR CODE HERE
# raise NotImplementedError()
```
